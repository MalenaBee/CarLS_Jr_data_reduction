```python
#FOR TESTING DATA REDUCTION TECHNIQUES

from SpecReduction import SpecFunctions as sf
import numpy as np
import pandas as pd
import scipy
import scipy.optimize as opt
from scipy.optimize import curve_fit
from astropy.io import fits
import matplotlib as mpl
import matplotlib.pyplot as plt
from aquarel import load_theme
```

```python
#style graphs
theme = load_theme("arctic_dark").set_font(family="monospace").set_grid(draw = False)
theme.apply()
```

```python
#average and display lights
lights_num = 10
lights_path = f'UNe_sat_lights_short/2025-07-29-2132_7-CapObj_'
lights_shape = fits.open(f'UNe_sat_lights_short/2025-07-29-2132_7-CapObj_0002.FIT')[0].data.shape
lights_avg = sf.median_img(lights_num, lights_path, lights_shape[0], lights_shape[1])

plt.figure(figsize = (12,12))
plt.imshow(lights_avg, norm = 'log')
plt.title('UNe Lights (Saturated) Stack Average')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.04)
plt.show()
```

```python
#average and display darks
darks_num = 10
darks_path = f'UNe_sat_darks_short/2025-07-29-2140_2-CapObj_'
darks_shape = fits.open(f'UNe_sat_darks_short/2025-07-29-2140_2-CapObj_0002.FIT')[0].data.shape
darks_avg = sf.median_img(darks_num, darks_path, darks_shape[0], darks_shape[1])

plt.figure(figsize = (12,12))
plt.imshow(darks_avg, norm = 'log')
plt.title('UNe Darks (Saturated) Stack Average')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.04)
plt.show()
```

```python
#average and display bias frames
bias_num = 5
bias_path = f'Bias729_short/2025-07-29-2238_7-CapObj_'
bias_shape = fits.open(f'Bias729_short/2025-07-29-2238_7-CapObj_0003.FIT')[0].data.shape
bias_avg = sf.median_img(bias_num, bias_path, bias_shape[0], bias_shape[1])

plt.figure(figsize = (12,12))
plt.imshow(bias_avg, norm = 'log')
plt.title('Bias Stack Average')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.04)
plt.show()
```

```python
lights_2 = lights_avg - bias_avg
darks_2 = darks_avg - bias_avg

final_UNe_sat = lights_2 - darks_2

plt.figure(figsize = (12,12))
plt.imshow(final_UNe_sat, norm = 'log')
plt.title('UNe (Saturated) Subtracted Image')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.04)
plt.grid(True)
plt.show()

```

```python
correct_UNe_sat = sf.straight_spec(final_UNe_sat)
#can't decide whether to use log scaling below
```

```python
#splice image into 1D spectrum
UNe_sat_1D = sf.make1D(correct_UNe_sat)


plt.plot(UNe_sat_1D)
plt.xlabel('x pixel')
plt.ylabel('normalized flux')
plt.title('1D Spectrum for UNe HCL (Saturated Run)')
plt.gca().invert_xaxis()
plt.show()
```

```python
#identify peaks
peaks, props = scipy.signal.find_peaks(UNe_sat_1D, height = 0.5)
loc = peaks
peak_val = np.array([])
edges = np.array([])
fwhm = scipy.signal.peak_widths(UNe_sat_1D, peaks, rel_height = 0.5)[0]
fw = scipy.signal.peak_widths(UNe_sat_1D, peaks, rel_height = 0.9)[0]

for p in range(peaks.size): #explore peak_widths function
    location = loc[p]
    peak_val = np.append(peak_val, UNe_sat_1D[location])
#     edges = np.append(edges, loc[p] - fwhm[p])
#     edges = np.append(edges, loc[p] + fwhm[p])

#plot
plt.figure(figsize = (8,6))
plt.plot(UNe_sat_1D, label = 'spectrum') #plots original data
plt.scatter(loc, peak_val, color = '#bf616a', label = 'peaks', lw = 3) #plots peaks identified above
# for edge in range(edges.size):
#     plt.axvline(edges[edge], color = 'b', lw = 0.1)
plt.legend()
plt.xlabel('x pixels')
plt.ylabel('normalized flux')
plt.title('Peaks in 1D UNe HCL (Saturated Run) Spectrum')
plt.xlim(5000, 5400)
plt.gca().invert_xaxis()
plt.show()

print(peaks.size)
```

```python
loc_clean = np.array([])
peak_val_clean = np.array([])
fwhm_clean = np.array([])
fw_clean = np.array([])

for peak in range(loc.size - 1):
    if loc[peak] + fwhm[peak] < loc[peak+1] - fwhm[peak+1] - 2:
        if peak == 0:
            loc_clean = np.append(loc_clean, loc[peak])
            peak_val_clean = np.append(peak_val_clean, peak_val[peak])
            fwhm_clean = np.append(fwhm_clean, fwhm[peak])
            fw_clean = np.append(fw_clean, fwhm[peak])
            continue
        elif loc[peak] - fwhm[peak] > loc[peak - 1] + fwhm[peak-1] + 2:
            loc_clean = np.append(loc_clean, loc[peak])
            peak_val_clean = np.append(peak_val_clean, peak_val[peak])
            fwhm_clean = np.append(fwhm_clean, fwhm[peak])
            fw_clean = np.append(fw_clean, fwhm[peak])

plt.figure(figsize = (8,6))
plt.plot(UNe_sat_1D, label = 'spectrum') #plots original data
plt.scatter(loc_clean, peak_val_clean, color = '#bf616a', label = 'peaks', lw = 2) #plots peaks identified above

plt.legend()
plt.xlabel('x pixels')
plt.ylabel('normalized flux')
plt.title('Peaks in 1D UNe HCL (Saturated Run) Spectrum')
# plt.xlim(7000, 7500)
plt.gca().invert_xaxis()
plt.show()
```

```python
pixels = np.linspace(0, int(UNe_sat_1D.size), int(UNe_sat_1D.size)+1)
x_center = np.array([])
for x in range(loc_clean.size):
    min_loc = int(loc_clean[x] - fw_clean[x])
    max_loc = int(loc_clean[x] + fw_clean[x])
    loc_range = pixels[min_loc:max_loc]
    i_range = UNe_sat_1D[min_loc:max_loc]
    num = 0
    for y in range(loc_range.size):
        num += loc_range[y]*i_range[y]
       
    x_center = np.append(x_center, num/np.sum(i_range))
    
plt.figure(figsize = (8,6))
plt.plot(UNe_sat_1D, label = 'spectrum') #plots original data
for x in range(x_center.size):
    plt.axvline(x_center[x], color = '#bf616a', lw = 1) #plots peaks identified above

plt.xlabel('x pixels')
plt.ylabel('normalized flux')
plt.title('Peaks in 1D UNe HCL (Saturated Run) Spectrum')
plt.ylim(0, 1)
plt.xlim(1350,1370)
plt.gca().invert_xaxis()
plt.show()
```

```python
residuals = np.array([])
for p in range(loc_clean.size):
    residual = loc_clean[p] - x_center[p]
    residuals = np.append(residuals, residual)

plt.scatter(np.arange(len(residuals)),residuals)
plt.title('Centroiding Residuals')
plt.xlabel('Peak Number')
plt.ylabel('Residual [peak value - centroid value]')
plt.axhline(0,ls='-.', c = '#bf616a')
plt.show()
```
