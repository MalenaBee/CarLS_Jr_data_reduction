```python
from SpecReduction import SpecFunctions as sf
import numpy as np
import pandas as pd
import scipy
import scipy.optimize as opt
from scipy.optimize import curve_fit
import specreduce
from specreduce import WavelengthCalibration1D as wav_cal
from astropy.io import fits
import matplotlib as mpl
import matplotlib.pyplot as plt
from aquarel import load_theme
from astropy import units as u
```

```python
theme = load_theme('arctic_dark')
theme.apply()
```

```python
#median of bias frames stack
bias_path = f'Lab_day_080525/Bias_85/2025-08-05_18_26_19Z/2025-08-05-1826_3-CapObj_'
bias_shape = fits.open(f'Lab_day_080525/Bias_85/2025-08-05_18_26_19Z/2025-08-05-1826_3-CapObj_0026.FIT')[0].data.shape
bias_avg = sf.median_img(10, bias_path, bias_shape[0], bias_shape[1])

plt.figure(figsize = (12,12))
plt.imshow(bias_avg)
plt.title('Bias Stack Average')
plt.xlabel('ThAr_pix')
plt.ylabel('ThAr_pix')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.08)
plt.show()
```

```python
#median of ThAr lights stack
ThAr_lights_path = f'Lab_day_080525/ThAr002_lights/2025-08-05_20_41_40Z/2025-08-05-2041_6-CapObj_'
ThAr_lights_shape = fits.open(f'Lab_day_080525/ThAr002_lights/2025-08-05_20_41_40Z/2025-08-05-2041_6-CapObj_0043.FIT')[0].data.shape
ThAr_lights_avg = sf.median_img(10, ThAr_lights_path, ThAr_lights_shape[0], ThAr_lights_shape[1])

ThAr_lights_avg = ThAr_lights_avg - bias_avg

plt.figure(figsize = (12,12))
plt.imshow(ThAr_lights_avg, norm = 'log')
plt.title('ThAr Lights Stack Average')
plt.xlabel('ThAr_pix')
plt.ylabel('ThAr_pix')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.08)
plt.show()
```

```python
#median of ThAr darks stack
ThAr_darks_path = f'Lab_day_080525/ThAr002_darks/2025-08-05_20_46_38Z/2025-08-05-2046_6-CapObj_'
ThAr_darks_shape = fits.open(f'Lab_day_080525/ThAr002_darks/2025-08-05_20_46_38Z/2025-08-05-2046_6-CapObj_0030.FIT')[0].data.shape
ThAr_darks_avg = sf.median_img(10, ThAr_darks_path, ThAr_darks_shape[0], ThAr_darks_shape[1])

ThAr_darks_avg = ThAr_darks_avg - bias_avg

plt.figure(figsize = (12,12))
plt.imshow(ThAr_darks_avg)
plt.title('ThAr Darks Stack Average')
plt.xlabel('ThAr_pix')
plt.ylabel('ThAr_pix')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.08)
plt.show()
```

```python
ThAr_subtracted = ThAr_lights_avg - ThAr_darks_avg
ThAr_flipped = np.fliplr(ThAr_subtracted)

plt.figure(figsize = (15,4))
plt.imshow(ThAr_subtracted, norm = 'log')
plt.title('ThAr Stack Average (GS002)')
plt.xlabel('Pixels')
plt.ylabel('Pixels')
plt.grid(True, c = 'white')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.2, norm = 'log')
# plt.savefig('ThAr85_02_stack.png')
plt.show()

ThAr02_85_hdul = fits.PrimaryHDU(ThAr_flipped)
ThAr02_85_hdul.writeto('FITS_shortcut/ThAr85_02_flipped.fits', overwrite = True)
```

```python
ThAr_corrected = sf.straight_spec(ThAr_flipped)
```

```python
plt.figure(figsize = (15,4))
plt.imshow(ThAr_corrected, norm = 'log', vmin = 1, vmax = 10**4.3)
plt.title('ThAr Rectified (GS002)')
plt.xlabel('Pixels')
plt.ylabel('Pixels')
plt.grid(True, c = 'white')
plt.colorbar(shrink = 0.4, location = 'bottom', pad = 0.2, norm = 'log')
plt.savefig('ThAr85_02_rectified.png')
plt.show()
```

```python
ThAr_spec_norm = sf.make1D(ThAr_corrected, False)
plt.plot(ThAr_spec_norm)
plt.title('ThAr GS002 Spectrum - Normalized')
plt.ylabel('Relative Intensity') #percentage of total intensity in each peak
plt.xlabel('Pixel (Wavelength Space)')
plt.xlim(0,9700)
plt.ylim(0, 1.1)
plt.savefig('ThAr85_02_spectrum.png')
plt.show()
```

```python
ThAr_pix = np.linspace(0, ThAr_spec_norm.size - 1, ThAr_spec_norm.size)

GS002_data = {'pixel': ThAr_pix,'intensity': ThAr_spec_norm}
GS002_df = pd.DataFrame(GS002_data)
GS002_df.to_csv('GS002.csv', index=False)
```
